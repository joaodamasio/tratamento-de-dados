## Criando Agupamento

```python
import pandas as pd
```

```python
dados = pd.read_csv('data/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
dados['Valor'].mean()
```

```python
bairro = dados['Bairro'].unique()
selecao = dados['Bairro'].isin(bairro)
dados = dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, dadin in grupo_bairro:
    print(f'{bairro}-> {dadin.Valor.mean()}')
```

```python
grupo_bairro[['Valor','Condominio']].mean().round(2)
```

## Estísticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min','max','sum']).rename(columns = {'min': 'Mínimo', 'max':'Máximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22} )
```
